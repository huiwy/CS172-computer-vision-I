```python
import utils
import BoW

from matplotlib import pyplot as plt
import matplotlib.image as mpimg 
import numpy as np
import cv2 as cv
```

```python
def read_image(f):
    img = mpimg.imread(f)
    t = img.shape
    if len(img.shape) == 3:
        img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    return img
                        
def show_image(data, label):
    plt.imshow(data, interpolation='nearest')
    plt.title(label)
    plt.show()
```

```python
im = read_image("170_0002.jpg")
```

```python
sift = cv.SIFT_create()
_, a = sift.detectAndCompute(im, None)
```

```python
show_image(im, "a")
```
